```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("MLDataset.csv",header = None)
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 291].values
```

```python
y.shape,x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
y_train.shape
```

```python
y_train = y_train.reshape((431,1))
```

```python
sc_X = StandardScaler()

x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)

sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```

# Using Decision Tree regressor

```python
clf = DecisionTreeRegressor(random_state=0)
clf.fit(x_train,y_train);
```

```python
y_pred = clf.predict(x_test)
y_pred = sc_y.inverse_transform(y_pred)
```

```python
#y_pred
```

```python
df_repre = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
```

```python
df1 = df_repre.head(20)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.ylabel("Salary")
plt.xlabel("Test case no")
plt.title("Predicted Salary values before Pruning")
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

# R2 SCORE OF DECISSION TREE REGRESSOR

```python
metrics.r2_score(y_test, y_pred)
```

# VISUALIZING DECESSION TREE

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True);
```

# Post pruning decision trees with cost complexity pruning

```python
path = clf.cost_complexity_pruning_path(x_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeRegressor(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(x_train, y_train)
    clfs.append(clf)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

# PLOTS FOR SELECTING ALPHA VALUES

```python
y_test_change = y_test.reshape((108,1))
y_test_change = sc_y.fit_transform(y_test_change)
```

```python
train_scores = [clf.score(x_train, y_train) for clf in clfs]

test_scores = [clf.score(x_test, y_test_change) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
train_scores = [clf.score(x_train, y_train) for clf in clfs]
test_scores = [clf.score(x_test, y_test_change) for clf in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
plt.xlim([0.0,0.01])
ax.legend()
plt.show()
```

```python
clf_acc = DecisionTreeRegressor(random_state=0, ccp_alpha = (0.008 + 0.0100)/2)
clf_acc.fit(x_train,y_train)
```

# R2 SCORE AFTER PRUNING

```python
y_pred=clf_acc.predict(x_test)
y_pred = sc_y.inverse_transform(y_pred)
metrics.r2_score(y_test, y_pred)
```

# VISUALIZING AFTER PRUNING

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf_acc,filled=True);
```
